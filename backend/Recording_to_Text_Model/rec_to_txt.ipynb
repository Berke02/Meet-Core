```python
import whisperx
import gc
import torch
import os
import pandas as pd
import torchaudio
from pyannote.audio import Pipeline
```

```python
from google.colab import userdata
token = userdata.get('HF_TOKEN')# Colab sol menü -> Secrets'a ekleyin
audio_file = "/content/new_clean_record.wav"
device = "cuda"
BATCH_SIZE = 16
COMPUTE_TYPE = "float16"
```

```python
model = whisperx.load_model("large-v3", device, compute_type=COMPUTE_TYPE, language="tr")#Specified the language because it could perform better

#Transcripting
audio = whisperx.load_audio(audio_file)
result = model.transcribe(audio, batch_size=BATCH_SIZE, language="tr")
```

```python
#Cleaning to not overload RAM and the VRAM
model = None
gc.collect()
torch.cuda.empty_cache()
```

```python
#alignment
model_a, metadata = whisperx.load_align_model(language_code=result["language"], device=device)
result = whisperx.align(result["segments"], model_a, metadata, audio, device, return_char_alignments=False)
```

```python
#one more cleaning
model_a = None
gc.collect()
torch.cuda.empty_cache()
```

```python
# diarization
pipeline = Pipeline.from_pretrained("pyannote/speaker-diarization-3.1", token=token)
pipeline.to(torch.device(device))

waveform, sample_rate = torchaudio.load(audio_file)
audio_input = {"waveform": waveform, "sample_rate": sample_rate}

diarize_output = pipeline(audio_input, min_speakers=2, max_speakers=5)

diarization = diarize_output.exclusive_speaker_diarization

diarize_segments = []
for segment, _, speaker in diarization.itertracks(yield_label=True):
    diarize_segments.append({
        "start": segment.start,
        "end": segment.end,
        "speaker": speaker
    })
diarize_df = pd.DataFrame(diarize_segments)

final_result = whisperx.assign_word_speakers(diarize_df, result)
```

```python
transcript_text = ""
limit = 0
for segment in final_result["segments"]:
    speaker = segment.get("speaker", "unknown")
    text = segment["text"].strip()
    start = segment["start"]
    time_str = f"{int(start // 60):02}:{int(start % 60):02}"

    satir = f"[{time_str}] {speaker}: {text}\n"
    if limit < 5:
        print(satir.strip())
        limit += 1
    transcript_text += satir

with open("final_t.txt", "w", encoding="utf-8") as f:
    f.write(transcript_text)
```
